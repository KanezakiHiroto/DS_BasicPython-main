## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
data1 = pd.read_csv('data/hist_data.csv')

plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度(℃)')
plt.ylabel('度数')

plt.xlim(0, 55)
plt.ylim(0, 30)

plt.hist(data1['aデータ'], bins=30, label='aデータ', alpha=0.5, color='orange')

plt.hist(data1['bデータ'], bins=30, label='bデータ', alpha=0.3, color='cyan')

plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
data2 = pd.read_csv('data/scatter_data.csv')

plt.title('クラス0～2のデータのlengthとwidthに関する散布図')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')

plt.xlim(4.0, 8.5)
plt.ylim(1.5, 4.5)

class0 = data2[data2['class'] == 0]
class1 = data2[data2['class'] == 1]
class2 = data2[data2['class'] == 2]

plt.scatter(class0['length (cm)'], class0['width (cm)'], label='class0', color='r')
plt.scatter(class1['length (cm)'], class1['width (cm)'], label='class1', color='b')
plt.scatter(class2['length (cm)'], class2['width (cm)'], label='class2', color='g')

plt.legend()
plt.grid()

plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
data3 = pd.read_csv('data/line_data.csv')
plt.title('期間ごとの件数の推移')
plt.xlabel('期間')
plt.ylabel('件数')
plt.xticks(rotation=30)
plt.ylim(100,150)
plt.grid()
plt.plot(data3['期間'], data3['件数'], marker='o')
plt.show()

plt.title('期間ごとの費用の推移')
plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.xticks(rotation=30)
plt.grid()
plt.bar(data3['期間'], data3['費用'], width=0.8)
plt.show()
```
